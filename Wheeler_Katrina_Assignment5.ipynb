```python
#Part 2
import numpy as np
import math
import random

np.random.seed(26)
greek_alpha = np.array(['Beta', 'Delta', 'Theta', 'Epsilon', 'Zeta', 'Gamma', 'Eta', 'Iota', 'Alpha'])

x_mean = 10
x_sd = 1.5
x_array = np.random.normal(loc=x_mean, scale=x_sd, size=9)

y_mean = 10
y_sd = 1.5
y_array = np.random.normal(loc=y_mean, scale=y_sd, size=9)

random.seed(26)
nine_elems = [1,2,3,4,5,6,7,8,9]
angle = [random.uniform(0, (2*math.pi)) for num in nine_elems]
cos_angle = np.cos(angle)

column_names = ['Letter', 'X', 'Y', 'Angle', 'Cosine']

dict_array = [{'Letter': a, 'X': b, 'Y':c, 'Angle':d, 'Cosine':e} for a,b,c,d,e in zip(greek_alpha, x_array, y_array, angle, cos_angle)]

greek_df = pd.DataFrame(dict_array)
print(greek_df)
print(" ")

final1 = greek_df.sort_values('Letter')
final2 = final1.drop(columns = ['X', 'Y'])
final3 = final2.drop(8)
print(final3)
```

```python
#Part 3
terms = 12
count = 0
n1 = 0
n2 = 1
fiblist = []
while count < terms:
    fiblist.append(n1)
    nth = n1 + n2
    n1 = n2
    n2 = nth
    count += 1
    
print(fiblist)
fib1 = fiblist[6:]
fib2 = fiblist[7:]
ratios = []
new = None

ratios = [(val1/val2) for val1,val2 in zip(fib1,fib2)]
print(ratios)
#The ratio iterates between increasing and decreasing compared to the number before it
```

```python
#Part 4
def convertToRankine(num):
        value = num * 1.8
        return value
    
kelvin_temps = (250, 300, 350, 400, 450)
for value in kelvin_temps:
    print(convertToRankine(value))
    
print(" ")
x = lambda a: a * 1.8
for value in kelvin_temps:
    print(x(value))
```

```python
# Part 5
col1 = np.random.normal(size = 100)
col2 = np.random.normal(size = 100)

d = {'Column1':col1, 'Column2':col2}
normal_df = pd.DataFrame(d)

part75 = normal_df.sample(frac = .75)
part25 = normal_df.drop(part75.index)

print('75% of the DataFrame:')
print(part75.reset_index(drop=True).head(12))
print(" ")
print('25% of the DataFrame:')
print(part25.reset_index(drop=True).head(12))
print(" ")
print("75% Mean & Standard Deviation\n")
print("Mean:\n",part75.mean(), "\n")
print("SD:\n", part75.std())
print(" ")
print("25% Mean & Standard Deviation\n")
print("Mean:\n",part25.mean(), "\n")
print("SD:\n",part25.std())
```
